```python
#How does the distribution of an amt differ across is_fraud categories?
plt.figure(figsize=(8,6))
#plt.xlim(0,500)
#plt.ylim(0,8000)
sns.displot(x='amt',data=df,hue='is_fraud')
plt.xlim(0,500)
plt.ylim(0,8000)
plt.plot()
```

```python
#Are there any outliers in the city_pop and amt?
sns.boxplot(x='city_pop',data=df)
```

```python
sns.boxplot(x='amt',data=df)
```

```python
#removal of outliers from amt and city_pop
columns = ['amt','city_pop']
for i in columns:
  q1 = df[i].quantile(0.25)
  q3 = df[i].quantile(0.75)
  iqr = q3-q1
  lower_bound = q1-1.5*iqr
  upper_bound = q3+1.5*iqr
  df = df[(df[i] >= lower_bound) & (df[i] <= upper_bound)]
```

```python
sns.boxplot(x='city_pop',data=df)
```

```python
df = df[df['city_pop']<10000]
sns.boxplot(x='city_pop',data=df)
```
